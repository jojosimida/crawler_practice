```python
import re
import requests
from bs4 import BeautifulSoup
from PIL import Image
from urllib.request import urlretrieve
```

```python
def check_percentage(chunk,size,remote):
    percent = 100.0 * chunk * size / remote
    if percent > 100.0:
        percent = 100.0
        
    print("Download...{:.2f}%".format(percent))
```

```python
def download_image(link):
    name = link.split("/")[-1]
    urlretrieve(link,"beauty_pic/"+name,check_percentage)
```

```python
url = "https://www.ptt.cc/bbs/Beauty/M.1506478888.A.F96.html"
response = requests.get(url)
soup = BeautifulSoup(response.text,"lxml")

pic = soup.find_all("a")
for i in pic:
    print(i.attrs)
    
    if re.search("i\.imgur", i["href"]):
        print()
        download_image(i["href"])
    
```
