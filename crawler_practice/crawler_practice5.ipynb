```python
import requests 
from bs4 import BeautifulSoup 
import re
import pandas as pd
import sys
```

```python
a = 'https://tw.movies.yahoo.com/movieinfo_review.html/id=6619'
response = requests.get(a)
response.encoding
```

```python
soup = BeautifulSoup(response.text,"lxml")
title = soup.find("div", class_='inform_title').text
title
```

```python
page = int(soup.find("div", class_="page_numbox").find_all("a")[-2].text)
page
```

```python
comment_df = pd.DataFrame(columns=["comments", "star"])
comments_all=[]
star_all = []
```

```python
for i in range(1,page):
    sys.stdout.write("\r正在抓取玩命關頭電影評論第 " + str(i) + "  頁")
    response = requests.get("https://tw.movies.yahoo.com/movieinfo_review.html/id=6619?sort=update_ts&order=desc&page=" +str(i))
    soup = BeautifulSoup(response.text,"lxml")
    
    comment = [i.find_next_sibling("span").text for i in soup.find_all("div",class_="user_star _c")]
    comments_all.extend(comment)
#   先從findall iterator 出每一項，再從每一項中找出他們的下一個兄弟並設條件span
#   為什麼這樣做是因為，評論的span沒有其他屬性，抓不到評論的文字
    
    star = [x['value'] for x in soup.find("ul", class_="usercom_list").find_all("input",{"name":"score"})]
    star_all.extend(star)

    
comment
```

```python
comment_df = pd.DataFrame({"comments":comments_all, "star":star_all,"movie":title})
print(len(comments_all))
print(len(star_all))
comments_all
```
