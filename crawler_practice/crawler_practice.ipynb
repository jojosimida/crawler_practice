```python
d = pd.read_html(url)[0]
l = d.iloc[1].tolist()

df = pd.DataFrame(columns=l)
ran = [0,2,3,4]


for i in ran:
    p = pd.read_html(url)[i]
    if i==0:
        p = p.drop(0,axis=0)
    
    old_names = p.columns.tolist()  #p.columns.values
    if i==0:
        new_names = p.loc[1].tolist()
    else:
        new_names = p.loc[0].tolist()

    p.rename(columns=dict(zip(old_names, new_names)), inplace=True)
   
    if i==0:
        p = p.drop(1,axis=0)
    else:
        p = p.drop(0,axis=0)
    

    for i in range(1,p.index.size):
        p.iloc[i] = p.iloc[i].shift(1)

    
    p['區域'] = p['區域'].fillna(method='ffill')
    df = pd.concat([df, p], axis=0, ignore_index=True)
    
df
```
